```python
# Point cloud Images for a small Dataset in ShapeNet
import os
import open3d as o3d

def read_point_cloud_files(path):
    """
    Function to read all point cloud files in a directory
    """
    pcds = []
    for file in os.listdir(path):
        if file.endswith(".ply"):
            pcd = o3d.io.read_point_cloud(os.path.join(path, file))
            pcds.append(pcd)
    return pcds

def visualize_point_cloud(pcds):
    """
    Function to visualize a list of point clouds
    """
    for pcd in pcds:
        o3d.visualization.draw_geometries([pcd])

if __name__ == "__main__":
    # Specify the path to your dataset directory
    dataset_path = "C:\\Users\\Archana Jayaraman\\Final_Project\\small_dataset"

    # Read all point cloud files in the dataset directory
    pcds = read_point_cloud_files(dataset_path)

    # Visualize the point clouds
    visualize_point_cloud(pcds)
```

```python
# This is the Pre-processing of the image. This will have missing components in output 
import open3d as o3d

def preprocess_and_visualize(input_file, voxel_size=0.01, outlier_nb_neighbors=20, outlier_std_ratio=2.0, subsample_every_k_points=10):
    # Load the .ply file
    ply_data = o3d.io.read_point_cloud(input_file)

    # Create a visualization window for both the original and preprocessed point clouds
    vis = o3d.visualization.Visualizer()
    vis.create_window()

    # Original point cloud (before preprocessing)
    vis.add_geometry(ply_data)
    vis.update_renderer()
    vis.poll_events()

    # Voxel grid downsampling
    downsampled_ply = ply_data.voxel_down_sample(voxel_size=voxel_size)

    # Statistical outlier removal
    filtered_ply, _ = downsampled_ply.remove_statistical_outlier(nb_neighbors=outlier_nb_neighbors, std_ratio=outlier_std_ratio)

    # Random subsampling
    subsampled_ply = filtered_ply.uniform_down_sample(subsample_every_k_points)

    # Preprocessed point cloud (after preprocessing)
    vis.clear_geometries()
    vis.add_geometry(subsampled_ply)
    vis.update_renderer()
    vis.poll_events()

    # Keep the visualization window open until the user closes it
    vis.run()

# Example usage:
input_file = "C:\\Users\\Archana Jayaraman\\OneDrive\\Desktop\\plyimage.ply"
preprocess_and_visualize(input_file)
```

```python
#Densify the point cloud/completion of point cloud 
import open3d as o3d
import numpy as np

def load_point_cloud(file_path):
    """Load a point cloud from a PLY file."""
    return o3d.io.read_point_cloud(file_path)

def densify_point_cloud(point_cloud, num_points=100000):
    """Densify the point cloud by interpolating new points."""
    # Estimate normals if not already estimated
    if not point_cloud.has_normals():
        point_cloud.estimate_normals()

    # Downsample the point cloud to speed up the process
    downsampled = point_cloud.voxel_down_sample(voxel_size=0.02)

    # Ensure that there are points in the downsampled point cloud
    if len(downsampled.points) == 0:
        print("The downsampled point cloud has no points.")
        return point_cloud

    # Create a KDTree for the downsampled point cloud
    kdtree = o3d.geometry.KDTreeFlann(downsampled)

    # Radius for the ball query
    radius = 0.05

    # Allocate new points
    new_points = np.asarray(point_cloud.points)
    while len(new_points) < num_points:
        # Randomly choose seed point for ball query
        seed_index = np.random.choice(len(downsampled.points))
        seed_point = np.asarray(downsampled.points)[seed_index]

        # Find neighboring points within the radius
        [k, idx, _] = kdtree.search_radius_vector_3d(seed_point, radius)

        # If neighbors found, interpolate a new point
        if k > 0:
            selected_points = np.asarray(downsampled.points)[idx, :]
            mean_point = np.mean(selected_points, axis=0)
            new_points = np.vstack((new_points, mean_point))

    # Create a new point cloud with the densified points
    new_point_cloud = o3d.geometry.PointCloud()
    new_point_cloud.points = o3d.utility.Vector3dVector(new_points[:num_points])
    return new_point_cloud

# Load point cloud
file_path = "C:\\Users\\Archana Jayaraman\\Final_Project\\small_dataset\\1b300fd9ad4050e6301fa0a0663ee996.ply" 
pcd = load_point_cloud(file_path)

# Densify the point cloud
densified_pcd = densify_point_cloud(pcd)

# Visualization
o3d.visualization.draw_geometries([densified_pcd])
```

```python
#preprocessing steps including data cleaning, downsampling, normalize color
#THis will run for all the ply files in the dataset. so pls make it for some random image
import open3d as o3d
import numpy as np
import glob
import os

def load_point_cloud(file_path):
    """Load a point cloud from a PLY file."""
    return o3d.io.read_point_cloud(file_path)

def remove_noise(point_cloud, nb_neighbors=20, std_ratio=2.0):
    """Remove noise using Statistical Outlier Removal."""
    clean_cloud, _ = point_cloud.remove_statistical_outlier(nb_neighbors, std_ratio)
    return clean_cloud

def downsample_point_cloud(point_cloud, voxel_size=0.05):
    """Downsample the point cloud using Voxel Grid Downsampling."""
    return point_cloud.voxel_down_sample(voxel_size)

def normalize_colors(point_cloud):
    """Normalize the colors of the point cloud."""
    colors = np.asarray(point_cloud.colors)
    colors = (colors - np.min(colors, axis=0)) / (np.max(colors, axis=0) - np.min(colors, axis=0))
    point_cloud.colors = o3d.utility.Vector3dVector(colors)
    return point_cloud

def visualize_point_cloud(point_cloud, title="Point Cloud"):
    """Visualize the point cloud."""
    print(title)
    o3d.visualization.draw_geometries([point_cloud])

# Directory where your PLY files are stored
directory_path = "C:\\Users\\Archana Jayaraman\\Final_Project\\small_dataset"  # Replace with your directory path

# Generate file paths for the PLY files in the directory
file_paths = glob.glob(os.path.join(directory_path, '*.ply'))

for file_path in file_paths:
    try:
        pcd = load_point_cloud(file_path)
        visualize_point_cloud(pcd, "Original Point Cloud")

        # Noise Removal
        pcd = remove_noise(pcd)
        visualize_point_cloud(pcd, "Point Cloud After Noise Removal")

        # Downsampling
        pcd = downsample_point_cloud(pcd)
        visualize_point_cloud(pcd, "Point Cloud After Downsampling")

        # Color Normalization
        if pcd.has_colors():
            pcd = normalize_colors(pcd)
            visualize_point_cloud(pcd, "Point Cloud After Color Normalization")

        # Save the preprocessed point cloud
        output_file = f"preprocessed_{os.path.basename(file_path)}"
        o3d.io.write_point_cloud(output_file, pcd)

        print(f"Preprocessed point cloud saved as {output_file}")

    except Exception as e:
        print(f"An error occurred while processing {file_path}: {e}")
```
